Q1. Load the given retail sales dataset into a Pandas DataFrame and display the first 5 rows along with the

dataset information to understand its structure.

```python
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv('Retail_Dataset.csv')

print("--- First 5 Rows ---")
print(df.head())

print("\n--- Dataset Information ---")
df.info()
```

Q2. Inspect the dataset and identify missing values in each column. Display the total number of missing

values present in the dataset.

```python
print("Missing values per column:")
missing_per_col = df.isnull().sum()
print(missing_per_col)

total_missing = df.isnull().sum().sum()
print(f"\nTotal number of missing values in the dataset: {total_missing}")
```

Q3. Replace missing values in the Quantity and Price columns with their respective mean values using

Pandas.

```python
# Calculate means
quantity_mean = df['Quantity'].mean()
price_mean = df['Price'].mean()

# Fill missing values
df['Quantity'] = df['Quantity'].fillna(quantity_mean)
df['Price'] = df['Price'].fillna(price_mean)

print("Missing values after imputation:")
print(df[['Quantity', 'Price']].isnull().sum())
```

Q4. Remove all rows where the Product Category or Region column contains missing values.

```python
# Remove rows where Product Category or Region is missing
df = df.dropna(subset=['Product Category', 'Region'])

print("Dataset shape after dropping missing categories:", df.shape)
```

Q5. Create a new column called Revenue using NumPy, where:



Revenue = Quantity × Price



Display the updated dataset with the new column.

```python
# Calculate revenue using NumPy
df['Revenue'] = np.multiply(df['Quantity'].values, df['Price'].values)

print("Updated dataset with Revenue column:")
print(df.head())
```

Q6. Using NumPy, calculate the total revenue generated from all transactions in the dataset.

```python
total_revenue = np.sum(df['Revenue'].values)
print(f"Total Revenue Generated: ${total_revenue:,.2f}")
```


Q7. Group the dataset by Product Category and calculate the total revenue generated by each category.

```python
category_revenue = df.groupby('Product Category')['Revenue'].sum().reset_index()
print("Revenue by Product Category:")
print(category_revenue)
```

Q8. Based on the revenue calculated for each product category, identify the top 3 categories with the

highest revenue and the bottom 3 categories with the lowest revenue.

```python
# Sorting categories by revenue
sorted_categories = category_revenue.sort_values(by='Revenue', ascending=False)

print("--- Top Categories ---")
print(sorted_categories.head(3))

print("\n--- Bottom Categories ---")
print(sorted_categories.tail(3))
```

Q9. Group the dataset by Region and calculate the total revenue generated in each region. Identify the

region with the highest revenue and the region with the lowest revenue.

```python
region_revenue = df.groupby('Region')['Revenue'].sum()
print("Revenue by Region:")
print(region_revenue)

highest_region = region_revenue.idxmax()
lowest_region = region_revenue.idxmin()

print(f"\nRegion with Highest Revenue: {highest_region} (${region_revenue[highest_region]:,.2f})")
print(f"Region with Lowest Revenue: {lowest_region} (${region_revenue[lowest_region]:,.2f})")
```

Q10. Extract the month from the Date column using Pandas and group the dataset by month to calculate the

total monthly revenue. Also, calculate the mean, median, and standard deviation of the Revenue column

using NumPy. solve this question answer this question fully explained answer tis question in googal collobrate



```python
# Convert Date to datetime and extract Month
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.to_period('M')

# Group by Month
monthly_revenue = df.groupby('Month')['Revenue'].sum().reset_index()
print("--- Total Monthly Revenue ---")
print(monthly_revenue)

# NumPy Statistical Summary
revenue_vals = df['Revenue'].values
rev_mean = np.mean(revenue_vals)
rev_median = np.median(revenue_vals)
rev_std = np.std(revenue_vals)

print("\n--- Revenue Statistical Summary ---")
print(f"Mean Revenue:   ${rev_mean:,.2f}")
print(f"Median Revenue: ${rev_median:,.2f}")
print(f"Std Dev:        ${rev_std:,.2f}")
```
